```python
numeric_columns = df.select_dtypes(include=['int', 'float']).columns
scaler = StandardScaler()
X_prep = scaler.fit_transform(df[numeric_columns])

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_prep)
```

```python
clusters = kmeans.predict(X_prep)
clusters
```

```python
pd.Series(clusters).value_counts().sort_index()
```

```python
X_df = df
X_df["cluster"] = clusters
X_df.head()

X_df['cluster'] = X_df['cluster'].apply(lambda x: 0 if x == 1 else 1 if x == 2 else 2)
X_df
```

```python
kmeans.inertia_
```

```python
K = range(2, min(20, len(X_prep) + 1))  
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=1234)
    kmeans.fit(X_prep)
    inertia.append(kmeans.inertia_)
    
inertia
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,11))
visualizer.fit(X_prep)
visualizer.show()
```

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(X_prep)
pca_centers = pca.transform(kmeans.cluster_centers_)
plt.figure(figsize=(10, 10))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
plt.scatter(pca_centers[:, 0], pca_centers[:, 1], c='red', s=200, alpha=0.5)
plt.title('KMeans Clusters')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=2)
dbscan.fit(X_prep)
df['cluster_dbscan'] = dbscan.labels_

df['cluster_dbscan'].value_counts()
```

```python
gmm = GaussianMixture(n_components=3)
gmm.fit(X_prep)
df['cluster_gmm'] = gmm.predict(X_prep)

df['cluster_gmm'].value_counts()
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=gmm.predict(X_prep), s=50, cmap='viridis')
plt.title('Gaussian Mixture Clusters')
plt.show()
```
